```python
from ultralytics import YOLO
# 모델 Load
model = YOLO("yolov8n.pt")

# 학습
results = model.train(
    data="/content/drive/MyDrive/Colab Notebooks/monkey.v1i.yolov8-hoo/data.yaml",
    epochs=100,
    imgsz=640
)
```

```python
# Validation 결과(mAP) 확인
metrics = model.val()
metrics.box.map
metrics.box.map50
metrics.box.map75
metrics.box.maps
```

```python
import glob
from PIL import Image

# 테스트 이미지 경로
test_img_folder = "/content/drive/MyDrive/Colab Notebooks/monkey.v1i.yolov8-hoo/test/images/"
test_imgs = glob.glob(test_img_folder + "*.jpg")

# 결과 저장 폴더
output_folder = "/content/drive/MyDrive/Colab Notebooks/monkey_test_results/"
import os
os.makedirs(output_folder, exist_ok=True)

# 예측 및 결과 이미지 저장
for img_path in test_imgs:
    results = model(img_path, save=True, project=output_folder)
    # Ultralytics가 자동으로 bounding box 그려서 저장함

print(f"Test 결과 이미지는 {output_folder} 폴더에 저장됩니다.")
```
